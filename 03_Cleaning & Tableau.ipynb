## Comp√©tences - simili - One Hot Encoder üõ†Ô∏è

```python
#Fonction pour prendre en charge les listes et les str dans le tableau
def normalize_skills(x):
    if isinstance(x, list):
        return [c.strip().lower() for c in x if isinstance(c, str) and c.strip()]
    elif isinstance(x, str):
        return [c.strip().lower() for c in x.split(",") if c.strip()]
    else:
        return []

# Normaliser la colonne comp√©tences
df["Comp√©tences (LLM)"] = df["Comp√©tences (LLM)"].apply(normalize_skills)

# Extraire toutes les comp√©tences uniques
all_skills = sorted({skill for skills in df["Comp√©tences (LLM)"] for skill in skills})
print(f"{len(all_skills)} comp√©tences uniques trouv√©es")

# Cr√©er colonnes binaires
for skill in all_skills:
    df[skill] = df["Comp√©tences (LLM)"].apply(lambda skills: 1 if skill in skills else 0)


df.head()
```

```python
csv_path = "data/suivi_candidatures_Tableau.csv"
df.to_csv(csv_path, sep=",", index=False, encoding="utf-8")
print(f"√âcrit: {csv_path}")
```

```python
# Tableau ne comprend pas le CSV parce qu'il est en fran√ßais et donc utilise ";" comme s√©parateur au lieu de "," 
# du coup des colonnes fant√¥mes not√©es F*** apparaissent et rendent impossible l'analyse

csv_path = "data/suivi_candidatures_Tableau_clean.csv"

# Tout convertir en string et nettoyer retours √† la ligne / tabulations
df_clean = df.astype(str).applymap(
    lambda x: x.replace("\n", " ").replace("\r", " ").replace("\t", " ")
)

# Sauvegarder en CSV avec point-virgule (plus s√ªr pour Tableau Public sur Mac FR)
df_clean.to_csv(
    csv_path,
    sep=";",
    index=False,
    encoding="utf-8",
    quoting=csv.QUOTE_ALL
)

print(f"‚úÖ CSV nettoy√© et √©crit : {csv_path}")
```
